```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Load the dataset
df = pd.read_csv("Debernardi et al 2020 data.csv")

# Handle missing values
imputer = SimpleImputer(strategy='mean')
df[['plasma_CA19_9', 'creatinine', 'LYVE1', 'REG1B', 'TFF1', 'REG1A']] = imputer.fit_transform(df[['plasma_CA19_9', 'creatinine', 'LYVE1', 'REG1B', 'TFF1', 'REG1A']])

# Check for outliers (you can adjust the threshold as needed)
outlier_threshold = 3
z_scores = (df[['plasma_CA19_9', 'creatinine', 'LYVE1', 'REG1B', 'TFF1', 'REG1A']] - df[['plasma_CA19_9', 'creatinine', 'LYVE1', 'REG1B', 'TFF1', 'REG1A']].mean()) / df[['plasma_CA19_9', 'creatinine', 'LYVE1', 'REG1B', 'TFF1', 'REG1A']].std()
df = df[(z_scores < outlier_threshold).all(axis=1)]

# Convert categorical variables to numerical representations
label_encoder = LabelEncoder()
df['sex'] = label_encoder.fit_transform(df['sex'])
df['diagnosis'] = label_encoder.fit_transform(df['diagnosis'])
df['stage'] = label_encoder.fit_transform(df['stage'])
df['benign_sample_diagnosis'] = label_encoder.fit_transform(df['benign_sample_diagnosis'])

# Scale numerical features
scaler = StandardScaler()
df[['plasma_CA19_9', 'creatinine', 'LYVE1', 'REG1B', 'TFF1', 'REG1A']] = scaler.fit_transform(df[['plasma_CA19_9', 'creatinine', 'LYVE1', 'REG1B', 'TFF1', 'REG1A']])

# Save cleaned dataset
df.to_csv("cleaned_dataset.csv", index=False)

```
